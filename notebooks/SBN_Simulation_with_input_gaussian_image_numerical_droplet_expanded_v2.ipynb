# <font color='green'>Create simulation initial conditions</font>

```python
def add_field_from_array_with_velocity_normalized_v2(array_to_add, x_af, y_af, Power_beam, intensity_normalization, 
                                                x0, y0, vx, vy, factor_t, normalization=True):
    
    d_x = x_af[1,0] - x_af[0,0]
    d_y = y_af[0,1] - y_af[0,0]
    
    norm = np.sum(np.abs(array_to_add)**2.0)*d_x*d_y/factor_t/factor_t*1e4

    I_beam = Power_beam/norm
    I_beam_normalized = I_beam/intensity_normalization
    
    if not normalization:
        new_field = af.interop.from_ndarray(array_to_add)*af.exp(1.0j*(vx*(x_af-x0)+ vy*(y_af-y0)))
    else:
        new_field = af.interop.from_ndarray(np.sqrt(I_beam_normalized)*array_to_add)*af.exp(1.0j*(vx*(x_af-x0)+ vy*(y_af-y0)))
    
    return af.to_array(new_field)

def droplet_2d_numerical(array_to_add, x_af, y_af, defect_power, defect_intensity_normalization, factor_t, normalization=True):
   
    d_x = x_af[1,0] - x_af[0,0]
    d_y = y_af[0,1] - y_af[0,0]
    
    norm = np.sum(np.abs(array_to_add)**2.0)*d_x*d_y/factor_t/factor_t*1e4
    
    I_beam = defect_power/norm
    I_beam_normalized = I_beam/defect_intensity_normalization
    if not normalization:
        new_field = af.interop.from_ndarray(array_to_add)   
    else:
        new_field = af.interop.from_ndarray(np.sqrt(I_beam_normalized)*array_to_add)
    return af.to_array(new_field)
```
